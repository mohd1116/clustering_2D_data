```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
import random

#data

def generate_random_numbers(seed, num_of_numbers):
    random.seed(seed)
    generated_numbers = []

    for _ in range(num_of_numbers):
        number = random.randint(0, 2**31 - 1)
        number_as_float = number / (2**31 - 1)
        generated_numbers.append(number_as_float)

    return generated_numbers

seed = 90141
data = generate_random_numbers(seed, 60)
newarr = np.array_split(data, 2)





# Combine x and y
df = np.column_stack((newarr[0],newarr[1]))

```

```python
print(df)
```

```python
# K=4
kmeans = KMeans(n_clusters=4, n_init=10, random_state=0).fit(df)
```

```python
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df[:, 0], df[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5, marker='X')
plt.title('K-Means Clustering for 2D Data')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```
